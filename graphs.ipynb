```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python


labels = ['1', '25', '50', '75', '100','125','150','175','200']
y = [0.015965,0.007999,0.007800,0.008137,0.008153,0.008115,0.007593,0.007602,0.007524]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, y, width, label='Loss')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('MSE Loss')
ax.set_xlabel('Number')
ax.set_title('Number of Neurons Analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



fig.tight_layout()

plt.show()
```
